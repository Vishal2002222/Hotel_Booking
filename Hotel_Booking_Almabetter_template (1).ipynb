# **Project Summary -**

### Project Summary: Hotel Booking Analysis

This project presents a comprehensive analysis of hotel booking data, focusing on two types of hotels: City Hotel and Resort Hotel. The objective was to uncover significant trends and insights that can inform strategic decisions for hotel management and marketing. The dataset encompasses various booking details, guest demographics, stay durations, and financial metrics.

#### Data Cleaning and Preprocessing

The initial dataset required substantial cleaning and preprocessing to ensure accuracy and relevance:

1. **Renaming Columns**: To enhance clarity, the 'adr' column was renamed to 'Average_Daily_Rate'.
2. **Handling Missing Values**: Missing values in key columns ('children', 'company', and 'country') were addressed. The 'children' column was filled with the mean value and converted to an integer. 'Company' was filled with zeros and its data type was adjusted. Missing values in the 'country' column were replaced with 'other'.
3. **Removing Duplicates**: The dataset contained 31,994 duplicate rows, which were removed to maintain data integrity.
4. **Filtering Invalid Data**: Rows where the total number of adults, children, and babies equaled zero were excluded, as they represent invalid bookings.

These steps ensured that the dataset was clean, with no null values and appropriate data types, ready for detailed analysis.

#### Exploratory Data Analysis (EDA)

The EDA phase aimed to identify patterns and generate insights:

1. **Booking Distribution**:
   - City Hotels accounted for 61% of the bookings, with Resort Hotels making up the remaining 39%.
   - Most guests were from Western European countries, indicating a regional preference.

2. **Cancellation Analysis**:
   - City Hotels experienced higher cancellation rates compared to Resort Hotels.
   - A longer lead time (the period between booking and check-in) correlated with a higher likelihood of cancellation.

3. **Booking Trends**:
   - Seasonal trends were evident in booking rates and average daily rates (ADR).
   - ADR peaked during the summer months (May to August), highlighting high demand.
   - Conversely, the lowest ADR was observed in November, December, February, and January, marking the off-peak season.

4. **Stay Duration**:
   - Short stays (less than 4 days) were more common in City Hotels.
   - Resort Hotels were preferred for longer stays, indicating different guest preferences.

5. **Distribution Channels**:
   - The majority of bookings (80%) were made through travel agents and tour operators (TA/TO).
   - Direct bookings and corporate channels were less utilized, presenting potential growth areas.

#### Key Insights and Recommendations

Several strategic recommendations emerged from the EDA:

1. **Target Market Focus**:
   - Allocate a significant portion of the marketing budget to Western European countries, the primary source of guests.

2. **Hotel Type Emphasis**:
   - Focus marketing efforts on City Hotels, which attract more bookings.
   - Enhance the appeal of Resort Hotels for longer stays through tailored packages and promotions.

3. **Seasonal Pricing Strategy**:
   - Implement dynamic pricing strategies to maximize revenue during peak seasons (summer months) and offer discounts during off-peak months to boost occupancy.

4. **Cancellation Mitigation**:
   - Develop policies and incentives to reduce cancellations, particularly for bookings with longer lead times.
   - Offer flexible booking options and clear cancellation policies to improve customer satisfaction.

5. **Enhancing Distribution Channels**:
   - Increase direct bookings through the hotelâ€™s website and loyalty programs.
   - Strengthen relationships with TA/TO while exploring opportunities to diversify distribution channels.

#### Conclusion

This analysis provides a detailed overview of booking trends and guest preferences, offering actionable insights to optimize hotel operations and marketing strategies. By addressing identified patterns and implementing the recommendations, hotels can improve booking rates, reduce cancellations, and enhance overall guest satisfaction. This strategic approach will help hotels to better allocate resources, target the right markets, and optimize pricing strategies throughout the year.

## ***1. Know Your Data***

1. **hotel**: Indicates the type of hotel (City Hotel or Resort Hotel).
2. **is_canceled**: Specifies whether the booking was canceled (1) or not (0).
3. **lead_time**: The number of days between the booking date and the arrival date.
4. **arrival_date_year**: The year of the arrival date.
5. **arrival_date_month**: The month of the arrival date.
6. **arrival_date_week_number**: The week number of the arrival date.
7. **arrival_date_day_of_month**: The day of the month of the arrival date.
8. **stays_in_weekend_nights**: The number of weekend nights stayed.
9. **stays_in_week_nights**: The number of weekday nights stayed.
10. **adults**: The number of adults included in the booking.
11. **children**: The number of children included in the booking.
12. **babies**: The number of babies included in the booking.
13. **meal**: The type of meal booked (e.g., Bed & Breakfast, Half board, etc.).
14. **country**: The country of origin of the guest.
15. **market_segment**: The market segment designation (e.g., Online TA, Offline TA/TO, etc.).
16. **distribution_channel**: The distribution channel through which the booking was made.
17. **is_repeated_guest**: Indicates whether the booking was made by a repeated guest (1) or not (0).
18. **previous_cancellations**: The number of previous bookings that were canceled by the guest.
19. **previous_bookings_not_canceled**: The number of previous bookings that were not canceled by the guest.
20. **reserved_room_type**: The room type originally reserved by the guest.
21. **assigned_room_type**: The room type assigned to the guest at check-in.
22. **booking_changes**: The number of changes made to the booking before arrival.
23. **deposit_type**: The type of deposit made for the booking.
24. **agent**: The ID of the travel agent that made the booking.
25. **company**: The ID of the company that made the booking.
26. **days_in_waiting_list**: The number of days the booking was on the waiting list before it was confirmed.
27. **customer_type**: The type of booking customer (e.g., Contract, Group, Transient, etc.).
28. **adr**: The average daily rate (the average revenue per room per day).
29. **required_car_parking_spaces**: The number of car parking spaces requested by the guest.
30. **total_of_special_requests**: The total number of special requests made by the guest.
31. **reservation_status**: The reservation status (e.g., Check-Out, Canceled, etc.).
32. **reservation_status_date**: The date at which the reservation status was updated.

This dataset contains comprehensive information about hotel bookings, guest demographics, booking details, and more, which can be analyzed to derive valuable insights for hotel management and marketing strategies.
